```python
# Import modules
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
import collections
import numpy as np
import pandas as pd
```

### Open train data and t'ekaaluk

```python
# Read train data in dataframe
train_df = pd.read_csv('data/train.csv')
# Check total number of records
len(train_df)
```

### Open validation data and t'ekaaluk

```python
# Read validation data in dataframe
valid_df = pd.read_csv('data/valid.csv')
# Check total number of records
len(valid_df)
```

```python
# Open dataframe
valid_df
```

### Random forest model

#### Reasons for neglection
**Time**: 'Time' has a greater range of value than other columns. It will affect the result. Also, it does not seem to have much meaning.  
**Amount**: Same as 'Time'.

```python
# Drop column list
drop_cols = ['Time', 'Amount']
# Drop columns
rtrain_df = train_df.drop(drop_cols, axis=1)
rvalid_df = valid_df.drop(drop_cols, axis=1)
```

```python
# Set parameters for random forest classifier
n_estimators = 200
class_weight = 'balanced'
max_depth = 10
# rf_model
rf_model = RandomForestClassifier(n_estimators=n_estimators, class_weight=class_weight, max_depth=max_depth)
```

```python
# Setup
X_train, y_train = rtrain_df.drop('Class', axis=1), rtrain_df.Class
X_valid, y_valid = rvalid_df.drop('Class', axis=1), rvalid_df.Class
```

```python
# Fit to model
rf_model.fit(X_train, y_train)
```

```python
# Get predictions
rf_model.predict(X_valid)
```

```python
# Get accuracy
rf_model.score(X=X_valid, y=y_valid)
```

### F1 score

```python
# Create a confusion matrix
metrics.confusion_matrix(y_valid, rf_model.predict(X_valid))
```

```python
# Get F1 score
metrics.f1_score(y_valid, rf_model.predict(X_valid))
```

### Test

```python
# Read test data in dataframe
test_df = pd.read_csv('data/test.csv')
# Check total number of records
len(test_df)
```

```python
# Open dataframe
test_df
```

```python
# Drop column list
drop_cols = ['Time', 'Amount']
# Drop columns
rtest_df = test_df.drop(drop_cols, axis=1)
```

```python
# Setup
X_test = rtest_df
```

```python
# Get predictions
pred = rf_model.predict(X_test)
pred
```

### Save result on result.csv

```python
# Read result data in dataframe
result_df = pd.read_csv('data/result.csv')
# Check total number of records
len(test_df)
```

```python
# Copy result_df
new_df = result_df.copy()
```

```python
# Add test prediction result (pred) to new_df
new_df['Prediction'] = pred 
new_df
```

```python
# Find how many predictions matches the result
new_df['Match'] = np.where(
    (new_df['Class'] == new_df['Prediction']),
    True,
    False
)
new_df
```

```python
# Count
collections.Counter(new_df.Match)
```

```python
# Find all the unmatched result
count = 0
for index, row in new_df.iterrows():
    if row['Match'] == False:
        count += 1
        print(index)
```

```python
# Save as .csv
new_df.to_csv('data/new_result.csv')
```
